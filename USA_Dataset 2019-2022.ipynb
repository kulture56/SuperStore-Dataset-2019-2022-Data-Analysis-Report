# USA Sales 2019-2022: Data Analysis Report

## Description
**The SuperStore dataset comprises a comprehensive sales record from a superstore, containing 9,994 entries across 19 distinct fields. The dataset includes order details, anonymized customer information, product specifics, and financial metrics. This report analyzes various aspects of the dataset to extract meaningful insights.**

# Features
The dataset includes the following columns:

* Order Details: order_id, order_date, ship_date
* Customer Information: customer
* Product Details: manufactory, product_name
* Segmentation: segment, category, subcategory
* Geographical Information: region, zip, city, state, country
* Financial Metrics: discount, profit, quantity, sales, profit_margin

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
# Set the theme for the plots
plt.style.use('dark_background')
```

```python
file_path = r'C:\Users\myloc\Downloads\superstore_dataset 2019-2022.csv'
df = pd.read_csv(file_path)
```

**Data Cleaning**

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True) 
```

**changing datatypes**

```python
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

**Aggregate Sales and Profit** 

```python
print("Total Sales:", round(df['sales'].sum(), 2))
```

```python
print("Total Profit:", round(df['profit'].sum(), 2))
```

**Average Profit Margin**

```python
print("Average Profit Margin:", round(df['profit_margin'].mean(), 2))
```

**Sales and Profit by Category**

```python
category_sales_profit = round(df.groupby('category')[['sales', 'profit']].sum().sort_values(by='sales', ascending=False), 2)
print(category_sales_profit)
```

**Sales and Profit per Region and Category**

```python
category_states_sales_profit = round(df.groupby(['category', 'region'])[['sales', 'profit']].sum().sort_values(by='category', ascending=False), 2)
print(category_states_sales_profit)
```

**Top 10 States by Sales**

```python
# Calculate total sales by state, sort in descending order, and take the top 10
top_states_sales = round(df.groupby('state')['sales'].sum().sort_values(ascending=False).head(10), 2)
print("Top 10 States by Sales:\n", top_states_sales)
```

**Top 10 States by Profit**

```python
top_states_sales = round(df.groupby('state')['profit'].sum().sort_values(ascending=False).head(10), 2)
print("Top 10 States by Profit:\n", top_states_sales)
```

**Top 10 States by Discount**

```python
top_states_sales = round(df.groupby('state')['discount'].sum().sort_values(ascending=False).head(10), 2)
print("Top 10 States by discount:\n", top_states_sales)
```

**Top 10 States by profit_margin**

```python
top_states_sales = round(df.groupby('state')['profit_margin'].sum().sort_values(ascending=False).head(10), 2)
print("Top 10 States by profit_margin:\n", top_states_sales)
```

**Top 10 Customers by State, Sales and Profit**

```python
top_customers = round(df.groupby(['customer', 'state'])[['sales', 'profit']].sum().sort_values(by='sales', ascending=False).head(10), 2)
print("Top 10 Customers by Sales and Profit:\n", top_customers)
```

**Regional Sales Analysis**

```python
region_sales = round(df.groupby('region')['sales'].sum().sort_values(ascending=False), 2)
print(region_sales)
```

**Correlation Analysis**

```python
correlation = round(df[['discount', 'profit', 'quantity', 'sales', 'profit_margin']].corr(), 2)
print("Correlation Matrix:")
print(correlation)
```

**Plot Heatmap of Correlation Matrix**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

## Visualizations 

**1. Plot Sales and Profit by Category**

```python
plt.figure(figsize=(10, 6))
category_sales_profit.plot(kind='bar', stacked=True)
plt.title('Total Sales and Profit by Category')
plt.xlabel('Category')
plt.ylabel('Amount')
plt.show()
```

**2.Pie chart for Sales by Category**

```python
plt.figure(figsize=(10, 5))
plt.pie(category_sales_profit['sales'], labels=category_sales_profit.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales by Category')
plt.show()
```

**3.Pie chart for Profit by Category**

```python
plt.figure(figsize=(10, 5))
plt.pie(category_sales_profit['profit'], labels=category_sales_profit.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Profit by Category')
plt.show()
```

**4. Plot Regional Sales**

```python
plt.figure(figsize=(10, 6))
region_sales.plot(kind='bar')
plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.show()
```

**5. Plot Monthly Sales Trend**

```python
df['month'] = df['order_date'].dt.to_period('M')
monthly_sales = df.groupby('month')['sales'].sum()

# Plot the monthly sales trend
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

**6. Sales by Category and Subcategory**

```python
plt.figure(figsize=(16, 9))
category_subcategory_sales = df.groupby(['category', 'subcategory'])['sales'].sum().unstack()

# Plot the stacked bar chart
category_subcategory_sales.plot(kind='bar', stacked=True, figsize=(16, 9))
plt.title('Sales by Category and Subcategory')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.legend(title='Subcategory')
plt.show()
```

**7. Total Quantity Sold by State**

```python
plt.figure(figsize=(14, 7))
state_quantity = df.groupby('state')['quantity'].sum().sort_values(ascending=False)
state_quantity.plot(kind='bar')
plt.title('Total Quantity Sold by State')
plt.xlabel('State')
plt.ylabel('Quantity Sold')
plt.show()
```

**8. Profit Margin by Category**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='profit_margin', data=df, palette='plasma')
plt.title('Profit Margin by Category')
plt.xlabel('Category')
plt.ylabel('Profit Margin')
plt.show()
```

**9. Distribution of Discounts**

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['discount'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Discounts')
plt.xlabel('Discount')
plt.ylabel('Frequency')
plt.show()
```

**10. Profit vs. Sales by Category**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sales', y='profit', hue='category', data=df, palette='viridis')
plt.title('Profit vs. Sales by Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend(title='Category')
plt.show()
```
